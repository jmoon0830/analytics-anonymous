```python
import pandas as pd
import csv 
import glob
import os
os.chdir('Resources')
```

```python
extension = 'csv'
ticker_files = [x for x in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in ticker_files ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')

combined_csv
```
