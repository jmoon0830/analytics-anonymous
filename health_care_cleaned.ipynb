```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
# Create a data frame from the combined_csv.csv
file_path = "Resources/combined_csv.csv"

file_df = pd.read_csv(file_path)

# Display data frame
file_df.head(15)
```

```python
# Sort the data frame by timestamp
sorted_df = file_df.sort_values(by=['timestamp'])

# Display data frame
sorted_df.head(15)
```

```python
# Dropping the dates before 2015-01-01 from the data frame
cleaned_df = sorted_df.loc[(sorted_df['timestamp']>'2014-12-31')]

# Display the data frame
cleaned_df.head(15)
```

```python
# Creating a data frame for our predicted results by dropping the timestamps between 2020-01-01 to present
predict_df = cleaned_df.loc[(cleaned_df['timestamp']<'2020-01-01')]

# Display data frame
predict_df.tail(15)
```

```python
# Creating a data frame for the health care stock data
health_care_df = predict_df.loc[predict_df['Sector'] == 'Health Care'].reset_index().drop(columns='index')

# Display data frame
health_care_df.head(15)
```

```python
# Finding the average stock price of health care stock based on the close price
avg_health_stock_df = pd.DataFrame(health_care_df.groupby(['timestamp'])['close'].mean()).reset_index()

x_values = avg_health_stock_df['timestamp']
y_values = avg_health_stock_df['close']

x_val = range(60)

# Perform a linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val, y_values)

# Create equation of line to calculate predicted 
regress_values = x_val * slope + intercept
line_eq = f'y = {round(slope,2)}x + {round(intercept,2)}'

fig, ax = plt.subplots(figsize=(20, 10))

# Create a scatter plot
plt.scatter(x_values, y_values, marker='o', facecolors='darkturquoise', edgecolors='black')

# Add the linear regression equation and line to plot on top of the scatter plot
plt.plot(x_values, regress_values, linewidth=3, color='darkviolet')
plt.annotate(line_eq, (0, 190), fontsize=20, color='darkviolet')

# Rotate tick marks
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
# Create a title and labels for x and y axis
plt.title('End of the Month vs. Average Closing Price (2015-2019)', fontsize=20)
plt.xlabel('Dates', fontsize=16)
plt.ylabel('Stock Price', fontsize=16)

# Save and show the graph

plt.show()
```

```python
# Get only the 2015 data
avg_health_stock_2015 = health_care_df.loc[(health_care_df['timestamp'] >= '2015-01-30') & 
                                                (health_care_df['timestamp'] < '2016-01-29')]

# Get the average price 
avg_health_stock_2015_grouped = avg_health_stock_2015.groupby(['timestamp']).mean().reset_index()

# Set the x and y values
x_values = avg_health_stock_2015_grouped['timestamp']
y_values = avg_health_stock_2015_grouped['close']

x_val = range(12)
# Perform a linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val, y_values)

# Create equation of line to calculate predicted 
regress_values = x_val * slope + intercept
line_eq = f'y = {round(slope,2)}x + {round(intercept,2)}'

fig, ax = plt.subplots(figsize=(20, 10))
# Create a handle for each plot
health_2015, = plt.plot(x_values, y_values, marker='o', markersize= 10, color='darkturquoise',
                        linewidth=3, label='2015')
line, = plt.plot(x_values, regress_values, linewidth=3, color='darkviolet')

plt.annotate(line_eq, (0, 142), fontsize=20, color='darkviolet')

# Set the legend to where the chart thinks is best
plt.legend(handles=[health_2015], loc="best")

plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
# Create title and labels for the x and y axis
plt.title('Average Health Care Stock Prices 2015', fontsize=20)
plt.xlabel('Dates', fontsize=16)
plt.ylabel('Price', fontsize=16)

plt.tight_layout()
plt.show()
```
